```python
import os
import glob
import pandas as pd
```

```python
# Use glob to match the pattern ‘csv’
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]

# Combine all files in the list and export as CSV
df = pd.concat([pd.read_csv(f) for f in all_filenames ])
df.reset_index(drop=True, inplace=True)

#export to csv
df.to_csv( "spotify_data.csv", index=False, encoding='utf-8-sig')
df.head(5)
```

```python
df2 = df.drop(['track', 'artist', 'uri'], axis=1)
df2.head(2)
```

```python
#Validation of distribution of output values
df2.groupby("target").count()
```

```python
#Validation of correlation between variables
import numpy as np 
df3=df2.corr()
df3
```

```python
#1. Sin key - crih
#2. Sin Tempo - chris  / - purvi
#3. Sin speechiness / purvi  -grecia
#4. Sin chorus_hit  / grecia 
#5. Sin key, Tempo, speechiness, chorus_hit- heidy
```

```python
# Sin chorus_hit y speechiness
data = df2.values
X = df2[['danceability', 'energy', 'key', 'loudness','mode', 'acousticness', 'instrumentalness', 'liveness','valence', 'tempo', 'duration_ms', 'time_signature', 'sections']]  
y = data[:, 15]
```

```python
# Reformat data
# data = df2.values
# X = data[:, 0:15]  
# y = data[:, 15]

print(data.shape, X.shape, y.shape)
data
# X[0]
# y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10)
```

## Data Preprocessing

```python
#Scale features (X) using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler(feature_range=(0,1)).fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test) 

X_train_scaled
```

One-hot encode the labels

```python
#One-hot encode output labels (y)
from tensorflow.keras.utils import to_categorical
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)

y_train_categorical
```

## Creating and defining our Deep Learning Model Architecture

```python
#Create a sequential model
from tensorflow.keras.models import Sequential
model = Sequential() 

from tensorflow.keras.layers import Dense
number_inputs = 13  

#Create hidden layers
model.add(Dense(units=13,activation='relu', input_dim=number_inputs))
model.add(Dense(units=120,activation='relu'))
model.add(Dense(units=80,activation='relu'))

#Create output layer
number_classes = 2
model.add(Dense(units=number_classes, activation='softmax')) 
```

Number of Hidden Nodes

```python
#Model Summary
model.summary() 
```

```python
#Compile the Model
model.compile(optimizer='nadam', loss='categorical_crossentropy', metrics=['accuracy'])
```

## Training the Model

```python
#Training the Model
import tensorflow as tf

model.fit(X_train_scaled,y_train_categorical,epochs=500, batch_size=2000, shuffle=True, verbose=2)
```

## Validation of the Model

```python
#Evaluate the Model using the testing data
model_loss, model_accuracy = model.evaluate(X_test_scaled, y_test_categorical, verbose=2)
    
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
